```python
df["InsuranceProvider"] = df["InsuranceProvider"].fillna("Unknown")
```

```python
df.dtypes
```

```python
date_cols = ["RegistrationDate","AdmissionDate","DischargeDate","BillingDate", 'SurgeryDate', 'FollowUpDate']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], format="%d-%m-%Y", dayfirst=True, errors="coerce")
```

```python
mask = df["SurgeryDate"].notna() & df["FollowUpDate"].isna()
df.loc[mask, "FollowUpDate"] = (
    df.loc[mask, "SurgeryDate"] + pd.Timedelta(days = 7)
)
```

```python
df.dtypes
```

CREATE NEW FEATURES

```python
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0, 18, 35, 50,65, 100],
    labels=["Child", "YoungAdult", "Adult", "Senior", "Elder"]
)
```

```python
df["HospitalStays"] = (df["DischargeDate"] - df["AdmissionDate"]).dt.days
```

```python
df["PaymentStatus"] = np.where(
    df["PendingAmount"] == 0,
    "Paid",
    "Pending"
)
```
