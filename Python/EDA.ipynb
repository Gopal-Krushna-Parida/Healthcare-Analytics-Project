Financial analysis

Total hospital revenue

```python
Total_revenue = df["TotalBillAmount"].sum()
print("Total Hospital Revenue :Rs.{:,.2f}".format(Total_revenue))
```

Average Bill

```python
Avg_Bill = df["TotalBillAmount"].mean()
print("Average_billAmount : Rs.{:,.2f}".format(Avg_Bill))
```

Highest Bill

```python
Highest_Bill = df["TotalBillAmount"].max()
print("Highest_BillAmount : Rs.{:,.2f}".format(Highest_Bill))
```

Lowest Bill

```python
Lowest_bill = df["TotalBillAmount"].min()
print("Lowest Bill Amount : Rs.{:,.2f}".format(Lowest_bill))
```

Pending amount

```python
Pending_Amount = df["PendingAmount"].sum()
print("Total Pending Amount : Rs.{:,.2f}".format(Pending_Amount))
```

Highest Pending Amount and details

```python
highest_pending_Amount = df.loc[df["PendingAmount"].idxmax()]
print(highest_pending_Amount[["PatientID", "PatientName", "PendingAmount"]])
```

Lowest Pending Amount

```python
lowest_pending_Amount = df.loc[df["PendingAmount"].idxmin()]
print(lowest_pending_Amount[["PatientID", "PatientName", "PendingAmount"]])
```

Revenue by Payment Mode

```python
Revenue = df.groupby("PaymentMode")["TotalBillAmount"].sum().reset_index(name="Total Amount")
Revenue["Total Amount"] = Revenue["Total Amount"].map("{:,.2f}".format)
print(Revenue)
```
